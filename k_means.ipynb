```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity

iris = load_iris()
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------------------
# 2. Fit K-Means Clustering
# ---------------------------------------
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

print("Cluster labels assigned to each sample:\n", labels)

# ---------------------------------------
# 3. Check similarity using cluster labels
# ---------------------------------------
sample1 = 0   # choose any sample index
sample2 = 10  # choose another

print("\nSample 1 cluster:", labels[sample1])
print("Sample 2 cluster:", labels[sample2])

if labels[sample1] == labels[sample2]:
    print("Samples are similar (same cluster).")
else:
    print("Samples are NOT similar (different clusters).")

# ---------------------------------------
# 4. Optional: Compute similarity score
# Using cosine similarity
# ---------------------------------------
similarity_score = cosine_similarity(
    X_scaled[sample1].reshape(1, -1),
    X_scaled[sample2].reshape(1, -1)
)[0][0]

print("\nCosine similarity score:", similarity_score)
```
